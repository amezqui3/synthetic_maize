```python
import os
import glob

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
src = '../2019_G2F_Shared/'
files = glob.glob(src + '*.xlsx') + glob.glob(src + '*.csv')
files = sorted(files)
files
```

```python
dfs = [None for i in range(len(files))]
bnames = []

for i in range(len(files)):
    _, fname = os.path.split(files[i])
    bname, ext = os.path.splitext(fname)
    bnames.append(bname)
    
    if ext == '.xlsx':
        dfs[i] = pd.read_excel(files[i])
    elif ext == '.csv':
        dfs[i] = pd.read_csv(files[i])
```

# Model data

```python
src = '../model/'
files = glob.glob(src + '*poly_model.csv')
files = sorted(files)
files
```

```python
dfs = [None for i in range(len(files))]
bnames = []

for i in range(len(files)):
    _, fname = os.path.split(files[i])
    bname, ext = os.path.splitext(fname)
    bnames.append('_' + bname[:9])
    
    if ext == '.xlsx':
        dfs[i] = pd.read_excel(files[i])
    elif ext == '.csv':
        dfs[i] = pd.read_csv(files[i])
    print(i, '\t', dfs[i].shape)
```

```python
for i in range(len(files)):
    if 'Unnamed: 0' in dfs[i].columns.values:
        dfs[i].drop(columns=['Unnamed: 0'], inplace=True)
    #dfs[i] = dfs[i][dfs[i]['curled']==0]
    if 'curled' in dfs[i].columns.values:
        mask = dfs[i]['curled'] == 1
        dfs[i].loc[mask, 'A2'] = np.nan
        dfs[i].loc[mask, 'N2'] = np.nan
        dfs[i].loc[mask, 'length'] = np.nan
        dfs[i].drop(columns=['curled'], inplace=True)
```

```python
colnames = [
    'End.Angle',
    'End.Length',
    'Mid.Angle',
    'Mid.Length',
    'Stem.Angle',
    'A1',
    'N1',
    'A2',
    'N2',
    'a',
    'b',
    'c',
    'd',
    'model_leaf_length']
```

```python
meta = list(dfs[0].columns[:8].values)
offset = 1
l1 = dfs[0 + offset].merge(dfs[2+offset], on=meta, how='outer', suffixes=[bnames[0 + offset], bnames[2 + offset]])
```

```python
j = 8
offset = len(colnames)
cols = l1.columns.values
for j in range(8, 8+offset):
    twice = [cols[j]] + [cols[j+offset]]
    colname = colnames[j-8]
    l1[colname] = l1[twice].mean(axis=1, skipna=True)
    l1.drop(columns=twice, inplace=True)
    print(j, twice, colname, sep='\t')
```

```python
print(l1.shape)
idx = l1.iloc[:, 8:].dropna(axis='index', how='any').index
l1 = l1.loc[idx]
print(l1.shape)
```

```python
filename = src + 'merged_l2_model_data.csv'
print(filename)
l1.to_csv(filename, index=False)
```
