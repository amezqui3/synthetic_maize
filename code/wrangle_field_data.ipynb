```python
import os
import glob

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
src = '../2019_G2F_Shared/'
files = glob.glob(src + '*.xlsx') + glob.glob(src + '*.csv')
files = sorted(files)
files
```

```python
dfs = [None for i in range(len(files))]
bnames = []

for i in range(len(files)):
    _, fname = os.path.split(files[i])
    bname, ext = os.path.splitext(fname)
    bnames.append(bname)
    
    if ext == '.xlsx':
        dfs[i] = pd.read_excel(files[i])
    elif ext == '.csv':
        dfs[i] = pd.read_csv(files[i])
```

```python
for i in range(len(files)):
    if 'Unnamed: 0' in dfs[i].columns.values:
        dfs[i].drop(columns=['Unnamed: 0'], inplace=True)
    if 'id' in dfs[i].columns.values:
        dfs[i].rename(columns={'id':'plot'}, inplace=True)
    if 'Plot' in dfs[i].columns.values:
        dfs[i].rename(columns={'Plot':'plot'}, inplace=True)
```

```python
for i in range(len(files)):
    print(i,'\t',dfs[i].shape, '\n', dfs[i].columns.values, '\n')
```

```python
meta = ['plot','New_Source','New_Pedigree','Old_Pedigree','Old_Source','Shared','New_Family','Tester']
drone_data = dfs[0].copy()
for i in range(1,6):
    drone_data = drone_data.merge(dfs[i], on=meta, how='inner', suffixes=['','_'+bnames[i]])
```

```python
new_name = []
old_name = drone_data.columns.values[len(meta):dfs[0].shape[1]]

for i in range(len(old_name)):
    new_name.append(old_name[i] + '_' + bnames[0])
new_name
```

```python
drone_data = drone_data.rename(columns=dict(zip(old_name, new_name)))
```

```python
for i,c in enumerate(drone_data.columns):
    print(i,c,sep='\t')
```

```python
i = 7
drone_data = drone_data.merge(dfs[i], on=meta, how='inner', suffixes=['','_'+bnames[i]])
```

```python
idx = drone_data.iloc[:, len(meta):].dropna(axis='index', how='any').index
drone_data = drone_data.loc[idx]
```

```python
meta2 = ['plot', 'Pass', 'Range']
pheno = dfs[8].merge(dfs[9], on=meta2, how='inner', suffixes=['_'+bnames[8], '_'+bnames[9]])
print(pheno.shape)
```

```python
meta3 = meta + meta2[1:] + ['Stand.Count', 'GreenSnap']
pheno = pheno.merge(dfs[6], on=meta3, how='inner', suffixes=['', '_'+bnames[6]])
print(pheno.shape)
```

```python
pheno
```

```python
data = pheno.merge(drone_data, on=meta, how='inner')
print(data.shape)
```

```python
cols = list(data.columns.values)
for i in range(len(cols)):
    print(i, '\t', cols[i])
```

```python
colss = cols[:11] + [cols[27]] + [cols[12]] + cols[16:21] + cols[40:42] + [cols[11]]
colss += cols[13:16] + cols[21:27] + cols[28:40] + cols[42:]
for i in range(len(colss)):
    print(i, '\t', colss[i])
```

```python
data = data[colss]
print(data.shape)
```

```python
idx = data.iloc[:, 20:].dropna(axis='index', how='any').index
data = data.loc[idx]
print(data.shape)
```

```python
idx = data['Quick.Note'][data['Quick.Note'].isna()].index
data = data.loc[idx]
print(data.shape)
```

```python
idx = data['X'][data['X'].isna()].index
data = data.loc[idx]
print(data.shape)
```

```python
data = data.drop(columns=['X', 'Quick.Note'])
print(data.shape)
```

```python
filename = src + 'merged_gxe_data.csv'
data.to_csv(filename, index=False)
```

# Model data

```python
src = '../model/'
files = glob.glob(src + '*poly_model.csv')
files = sorted(files)
files
```

```python
dfs = [None for i in range(len(files))]
bnames = []

for i in range(len(files)):
    _, fname = os.path.split(files[i])
    bname, ext = os.path.splitext(fname)
    bnames.append(bname[:9])
    
    if ext == '.xlsx':
        dfs[i] = pd.read_excel(files[i])
    elif ext == '.csv':
        dfs[i] = pd.read_csv(files[i])
```

```python
for i in range(len(files)):
    if 'Unnamed: 0' in dfs[i].columns.values:
        dfs[i].drop(columns=['Unnamed: 0'], inplace=True)
    dfs[i] = dfs[i][dfs[i]['curled']==0]
    if 'curled' in dfs[i].columns.values:
        dfs[i].drop(columns=['curled'], inplace=True)
```

```python
meta = list(dfs[0].columns[:8].values)

model_data = dfs[0].copy()
for i in range(1,len(dfs)):
    model_data = model_data.merge(dfs[i], on=meta, how='inner', suffixes=['','_'+bnames[i]])
```

```python
new_name = []
old_name = model_data.columns.values[len(meta):dfs[0].shape[1]]

for i in range(len(old_name)):
    new_name.append(old_name[i] + '_' + bnames[0])
new_name
```

```python
model_data = model_data.rename(columns=dict(zip(old_name, new_name)))
```

```python
model_data
```

```python
filename = src + 'merged_model_data.csv'
model_data.to_csv(filename, index=False)
```
