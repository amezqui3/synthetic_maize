```python
import rawpy
import imageio
import os
import glob
from skimage.color import rgb2gray
```

```python
src = '../maize/'
dst = '../stalk/'

raw_pics = glob.glob(src + '*.ARW')
```

```python
pic = raw_pics[25]
with rawpy.imread(pic) as raw:
    rgb = raw.postprocess()

_, fname = os.path.split(pic)
bname = (os.path.splitext(fname)[0])[1:]
bname
```

```python
gray = (rgb2gray(rgb)*255).astype('uint8')
```

```python
img = gray.copy()
thrh = 150
thrl = 75

r = rgb[:, :, 0] 
g = rgb[:, :, 1] 
b = rgb[:, :, 2]

maskh = (r > thrh)*(g > thrh)*(b > thrh*0.75)
maskl = (r < thrl*.75)*(g < thrl)*(b < thrl*1.25)

img[maskh] = 0
img[maskl] = 0
```

```python
imageio.imsave(dst + bname + '.jpg', img, format='jpg', optimize=True)
```

```python
import skimage.graph
### give start (y1,x1) and end (y2,x2) and the binary maze image as input
def shortest_path(start,end,binary):
    costs=np.where(binary,1,1000)
    path, cost = skimage.graph.route_through_array(
        costs, start=start, end=end, fully_connected=True)
    return path,cost
```

```python
np.argmax(box[0, :])
np.argmax(box[-1, :])
```

```python
path, cost = shortest_path([0,np.argmax(box[0,:])], [-1, np.argmax(box[-1,:])], box)
path = np.array(path)
```

```python
import pylab as pl
import numpy as np
import cv2

def collapse_contours(contours):
    contours = np.vstack(contours)
    contours = contours.squeeze()
    return contours
```

```python
contours, hierarchy = cv2.findContours(skeleton.astype('uint8'), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour = collapse_contours(contours[0])
```
